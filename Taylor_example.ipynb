## $f(x)=\sin(x)+\cos(x)$ のテイラー展開（マクローリン展開）

\begin{align*}
      \sin(x) + \cos(x)
      &= x -  \frac{1}{3!} x^3 + \frac{1}{5!} x^5 + \cdots + \frac{(-1)^{k}}{(2k+1)!} x^{2k+1} + \cdots \\
      &{} + 1 -  \frac{1}{2!} x^2 + \frac{1}{4!} x^4 + \cdots + \frac{(-1)^{k}}{(2k)!} x^{2k} + \cdots \\
      &= \sum_{k=0}^{\infty} (-1)^{k}\frac{1}{(2k)!} x^{2k} + \sum_{k=0}^{\infty} (-1)^{k} \frac{1}{(2k+x)!} x^{2k+1} \\
      &= 1 + x - \frac{1}{2!} x^2 - \frac{1}{3!} x^{3} + \frac{1}{4!} x^4 + \frac{1}{5!}x^5 - \cdots
\end{align*}


```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f(x):
    return np.sin(x)+np.cos(x)
```

```python
def t0(x):
  y = np.zeros(len(x)) + 1.0
  return y
```

```python
def t1(x):
  y = 1.0 + x
  return y
```

```python
def t5(x):
  y = 1.0 + x - np.power(x, 2)/math.factorial(2) - np.power(x, 3)/math.factorial(3) + np.power(x, 4)/math.factorial(4) + np.power(x, 5)/math.factorial(5)
  return y
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = f(x)
y0 = t0(x)
y1 = t1(x)
y5 = t5(x)

#print(x)
#print(y)
#print(y0)
#print(y5)

plt.title('Taylor Expansion')
plt.plot(x, y, label='f(x)=sin(x)+cos(x)')
plt.plot(x, y0, label='T0')
plt.plot(x, y1, label='T1')
plt.plot(x, y5, label='T5')
plt.legend()
plt.savefig('taylor_example.png')
plt.show()
```
