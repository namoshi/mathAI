# スチューデントの$t$ 分布

\begin{align*}
 f(x,k) = \frac{\Gamma((\nu + 1)/2}{\sqrt{\nu \pi} \Gamma(\nu/2)} (1+x^2/\nu)^{-(1+\nu)/2}
\end{align*}

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

for i in range(5):
  df = i+1
  x = np.linspace(-4.0, 4.0, 100)
  label='df={}'.format(df)
  plt.plot(x, t.pdf(x, df), '-', label=label)

plt.title('t Distribution')
plt.legend()
plt.savefig('t-distribution.png')
plt.show()
```
