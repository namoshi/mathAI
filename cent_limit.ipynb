```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(10000)

plt.hist(X, bins=30, range=(0.0, 1.0))
plt.show()
```

```python
X1 = np.random.rand(10000)
X2 = np.random.rand(10000)

X = (X1 + X2)/2

plt.hist(X, bins=30, range=(0.0, 1.0))
plt.show()
```

```python
n= 5
X=np.zeros(10000)
for i in range(n):
  X = X + np.random.rand(10000)

X=X/n
plt.hist(X, bins=30, range=(0.0, 1.0))
plt.show()
```

```python
n= 10
X=np.zeros(10000)
for i in range(n):
  X = X + np.random.rand(10000)

X=X/n
plt.hist(X, bins=50, range=(0.0, 1.0))
plt.show()
```
