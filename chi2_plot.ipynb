# $\chi^2$ 分布

\begin{align*}
 f(x,k) = \frac{1}{2^{k/2} \Gamma(k/2)} x^{k/2-1} \exp(-x/2)
\end{align*}

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

for i in range(5):
  df = i+1
  x = np.linspace(0.0, 10.0, 100)
  label='df={}'.format(df)
  plt.plot(x, chi2.pdf(x, df), '-', label=label)

plt.title('Chi Square Distribution')
plt.legend()
plt.savefig('chi-square-distribution.png')
plt.show()
```
